```python
# 1 Create an ElectricCar class that inherits from the Car class and includes an additional attribute, battery_size.
```

```python
class Car:
    def __init__(self):
        self.__speed = 0

    def increase_speed(self, amount):
        if amount > 0:
            self.__speed += amount

    def decrease_speed(self, amount):
        if amount > 0:
            self.__speed -= amount
            if self.__speed < 0:
                self.__speed = 0

    def get_speed(self):
        return self.__speed


# Child class
class ElectricCar(Car):
    def __init__(self, battery_size):
        super().__init__()   # call parent constructor
        self.battery_size = battery_size   # new attribute

    def get_battery_info(self):
        return f"Battery Size: {self.battery_size} kWh"


# Example usage
e_car = ElectricCar(75)

e_car.increase_speed(80)
print("Speed:", e_car.get_speed())

print(e_car.get_battery_info())
```

```python
#2 Define a Laptop class that inherits from the Computer class and includes an additional attribute, battery_life.
```

```python
# Parent class
class Computer:
    def __init__(self, brand, processor):
        self.brand = brand
        self.processor = processor

    def get_info(self):
        return f"Brand: {self.brand}, Processor: {self.processor}"


# Child class
class Laptop(Computer):
    def __init__(self, brand, processor, battery_life):
        super().__init__(brand, processor)   # call parent constructor
        self.battery_life = battery_life     # additional attribute

    def get_battery_info(self):
        return f"Battery Life: {self.battery_life} hours"


# Example usage
laptop = Laptop("Dell", "Intel i5", 8)

print(laptop.get_info())
print(laptop.get_battery_info())
```

```python
#3 Create a Smartphone class and add another child class of your choice that inherits from an ElectronicDevice class. Include an additional attribute, operating_system.
```

```python
# Parent class
class ElectronicDevice:
    def __init__(self, brand):
        self.brand = brand

    def get_brand(self):
        return f"Brand: {self.brand}"


# Child class 1
class Smartphone(ElectronicDevice):
    def __init__(self, brand, operating_system):
        super().__init__(brand)
        self.operating_system = operating_system

    def get_info(self):
        return f"{self.brand} Smartphone with {self.operating_system}"


# Child class 2 (example: Tablet)
class Tablet(ElectronicDevice):
    def __init__(self, brand, operating_system):
        super().__init__(brand)
        self.operating_system = operating_system

    def get_info(self):
        return f"{self.brand} Tablet with {self.operating_system}"


# Example usage
phone = Smartphone("Samsung", "Android")
tablet = Tablet("Apple", "iOS")

print(phone.get_info())
print(tablet.get_info())
```
