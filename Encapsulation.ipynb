```python
#Question 2:
#Create a BankAccount class with a private attribute __balance. Implement methods to deposit, withdraw, and check balance while ensuring the balance cannot be modified directly.
```

```python
class BankAccount:
    def __init__(self, balance=0):
        self.__balance = balance   # private attribute

    # Method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: {amount}")
        else:
            print("Invalid amount")

    # Method to withdraw money
    def withdraw(self, amount):
        if amount > self.__balance:
            print("Insufficient balance")
        elif amount > 0:
            self.__balance -= amount
            print(f"Withdrawn: {amount}")
        else:
            print("Invalid amount")

    # Method to check balance
    def get_balance(self):
        return self.__balance


# Example usage
account = BankAccount(1300)

account.deposit(500)
account.withdraw(330)

print("Current Balance:", account.get_balance())
```

```python
#Question 3:
#Create a Movie class with a private attribute __rating. Implement methods to set and get the rating, ensuring the rating is between 1 and 10.
```

```python
class Movie:
    def __init__(self):
        self.__rating = None   # private attribute

    # Method to set rating
    def set_rating(self, rating):
        if 1 <= rating <= 10:
            self.__rating = rating
            print(f"Rating set to {rating}")
        else:
            print("Invalid rating! Must be between 1 and 10.")

    # Method to get rating
    def get_rating(self):
        return self.__rating


# Example usage
movie = Movie()

movie.set_rating(8)
print("Movie Rating:", movie.get_rating())

movie.set_rating(15)   # Invalid case
```

```python
#Question 4:
#Create a Book class with private attributes __title and __author. Implement a constructor to set these values and provide methods to retrieve them.
```

```python
class Book:
    def __init__(self, title, author):
        self.__title = title     # private attribute
        self.__author = author   # private attribute

    # Method to get title
    def get_title(self):
        return self.__title

    # Method to get author
    def get_author(self):
        return self.__author


# Example usage
book = Book("Atomic Habits", "James Clear")

print("Title:", book.get_title())
print("Author:", book.get_author())
```

```python
#Question 5:(Optional)
#Create a Car class with a private attribute __speed. Implement methods to increase speed, decrease speed, and get the current speed, ensuring speed does not go below 0.
```

```python
class Car:
    def __init__(self):
        self.__speed = 0   # private attribute

    # Method to increase speed
    def increase_speed(self, amount):
        if amount > 0:
            self.__speed += amount
            print(f"Speed increased by {amount}")
        else:
            print("Invalid amount")

    # Method to decrease speed
    def decrease_speed(self, amount):
        if amount > 0:
            self.__speed -= amount
            if self.__speed < 0:
                self.__speed = 0   # ensure speed doesn't go below 0
            print(f"Speed decreased by {amount}")
        else:
            print("Invalid amount")

    # Method to get current speed
    def get_speed(self):
        return self.__speed


# Example usage
car = Car()

car.increase_speed(80)
car.decrease_speed(20)
car.decrease_speed(50)   # will not go below 0

print("Current Speed:", car.get_speed())
```
